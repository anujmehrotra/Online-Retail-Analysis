```python
most_freq=df1[["StockCode","Description"]].value_counts().reset_index()
most_freq
```

```python
most_freq[most_freq.StockCode=="85123A"].head(1)
```

```python
most_freq=most_freq.groupby("StockCode").head(1)
most_freq
```

```python
most_freq.columns=["StockCode","freq_Description","count"]

df2=df1.merge(most_freq,on="StockCode",how="left")
df2.head()
```

```python
df2['Description']=df2["freq_Description"]
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2.dropna(subset=["Description"],inplace=True)
df2.isnull().sum()
```

```python
df2.drop(columns=["freq_Description","count"],inplace=True)
df2.head()
```

```python
df2.describe()
```

```python
df2[df2.UnitPrice<0]
```

```python
df3=df2[(df2.Quantity>0) & (df2.UnitPrice>0)]
df3.head()
```

#### Feature Engineering:Create New Columns

```python
df4=df3.copy()
df4["TotalSales"]=df4["Quantity"]*df4["UnitPrice"]
df4.head(3)
```

```python
df4.info()
```

```python
df4['Month']=df4["InvoiceDate"].dt.month
df4.head(3)
```

#### Visualization & EDA

##### 1.Plot Monthly Sales

```python
monthly_sales=df4.groupby("Month").TotalSales.sum()
monthly_sales.plot(kind="line",title="Monthly Sales",marker="o")
plt.xlabel("Month")
plt.ylabel("TotalSales")
plt.grid()
plt.show()
```

####**Insights**
#### Total Sales started rising up in August having a peek in November.This is likely due to the holiday season at the end of the year


#### 2. Top 5 Countries

```python
top_5_countries=df4.groupby('Country')['TotalSales'].sum().sort_values(ascending=False).head(5)
top_5_countries
```

```python
top_5_countries.plot(kind='barh',title='Top 5 countries by Sales')
plt.xlabel('TotalSales')
plt.ylabel('Country')
plt.grid(axis='x')
plt.show()
```

```python
total_sales_all_countries = df4['TotalSales'].sum()
top_5_countries_percentage = (top_5_countries / total_sales_all_countries) * 100

ax = top_5_countries_percentage.plot(kind='barh', title='Top 5 Countries by Sales Percentage')
plt.xlabel('Percentage of Total Sales')
plt.ylabel('Country')
plt.grid(axis='x')

# Add percentage values to the bars
for i, v in enumerate(top_5_countries_percentage):
    ax.text(v + 1, i, f'{v:.1f}%', va='center')

plt.show()
```

```python
product_sales = df4.groupby('StockCode')['TotalSales'].sum()

# Reverse the order to display in descending order on the horizontal bar chart
product_sales = product_sales.iloc[::-1]
top_5_products=product_sales.sort_values(ascending=False).head(5)
top_5_products.plot(kind='barh', title='Product Sales',color='skyblue')
plt.xlabel('Total Sales')
plt.ylabel('Stock Code')
plt.gca().invert_yaxis()
plt.show()
```

#### 4.RFM Analysis

```python
current_dt=df4["InvoiceDate"].max()+pd.Timedelta(days=1)
current_dt
```

```python
rfm=df4.groupby("CustomerID").agg({
    "InvoiceDate":lambda x:(current_dt-x.max()).days,
    "InvoiceNo":"count",
    "TotalSales":"sum"
})

rfm.columns=["Recency","Frequency","Monetary"]
rfm.head(5)
```

```python
df4[df4.CustomerID==12347.0]["TotalSales"].sum()
```

```python
rfm['R_score']=pd.qcut(rfm["Recency"],4,labels=[4,3,2,1])
rfm['F_score']=pd.qcut(rfm["Frequency"],4,labels=[1,2,3,4])
rfm['M_score']=pd.qcut(rfm["Monetary"],4,labels=[1,2,3,4])
rfm['RFM_Score']=rfm[['R_score','F_score','M_score']].sum(axis=1)
rfm
```

```python
rfm.sort_values(by="RFM_Score",ascending=False)
```

#### 5. Customer Churn Analysis

```python
df4.head(2)
```

```python
customer_last_purchase=df4.groupby("CustomerID")['InvoiceDate'].max()
customer_last_purchase.head(5)
```

```python
type(customer_last_purchase)
```

```python
customer_last_purchase=(current_dt-customer_last_purchase).dt.days
customer_last_purchase.head(5)
```

```python
churn_threshold=90
churned_customers=customer_last_purchase[customer_last_purchase>churn_threshold]
churned_customers.head(5)
```

```python
print("Number of Churned Customers:",len(churned_customers))
plt.figure(figsize=(10,6))
plt.hist(customer_last_purchase,bins=50,color='red',alpha=0.7)
plt.axvline(churn_threshold,color='black',linestyle='dashed',linewidth=2)
plt.title('Customer Churn Distribution')
plt.xlabel('Days Since Last Purchase')
plt.ylabel('Number of Customers')
plt.show()
```
